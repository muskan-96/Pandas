## 3.Filling the Missing Values-Imputation
Imputation means filling in missing data in your dataset with some reasonable values instead of leaving them blank or removing those rows
1)Numerical data-use mean

```python
df3=df.copy()
```

```python
df3.head()
```

```python
mean_value=df3["reviews_per_month"].mean()
df3["reviews_per_month"]=df3["reviews_per_month"].fillna(mean_value)
#we are taking the mean value and filling it in null places 
```

```python
df3.head()
```

```python
df3.isnull().sum()
```

## Filling categorical data

```python
df3.nunique()
#df.nunique() returns a Series, where each entry shows the count of unique values for that column
```

```python
most_common = df3["last_review"].value_counts().index[0]
df3["last_review"] = df3["last_review"].fillna(most_common)
```

```python
df3.isna().sum()
```

```python
df4=df.copy()
df4["last_review"]
```

```python
df4.isna().sum()
```

```python
df4["last_review"]=df4["last_review"].fillna("Not_Revieved")
#fillna is used to fill the values 
```

```python
df4
```

## Advanced Imputation 

```python
df5=df.copy()
```

```python
df5.isna().sum()
```

```python
df5["reviews_per_month"]=df5["reviews_per_month"].interpolate()
#The interpolate() method is used to fill in missing values (NaNs) in a smart,
#mathematical way—by estimating them based on other values in the column.
#It "guesses" missing values by using values before and after the gap
```
