```python
import random
import matplotlib.pyplot as plt

# Simulate rolling a dice 1000 times
rolls = 1000
dice_results = []

for _ in range(rolls):
    dice_results.append(random.randint(1,6))  # random number 1-6

# Count how many times each number appeared
counts = {i: dice_results.count(i) for i in range(1,7)}
#for i in range(1,7) ‚Üí loop i = 1, 2, 3, 4, 5, 6
#dice_results.count(i) ‚Üí counts how many times i appears in dice_results
#{i: dice_results.count(i)} ‚Üí makes a dictionary entry for this number

# Calculate probabilities
probabilities = {num: count/rolls for num, count in counts.items()}

print("Counts:", counts)
print("Probabilities:", probabilities)

# Plotting
numbers = list(probabilities.keys())
probs = list(probabilities.values())

plt.bar(numbers, probs, color="skyblue")
plt.xlabel("Dice Number")
plt.ylabel("Probability")
plt.title("Probability of Dice Rolls (Simulated)")
plt.show()
```
