```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("AB_NYC_2019.csv")
df.head(2)
```

# Histogram

```python
sns.histplot(data=df,x="price")
```

```python
sns.histplot(data=df[df["price"]<500],x="price")
```
