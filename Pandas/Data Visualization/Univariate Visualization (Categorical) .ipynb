```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("AB_NYC_2019.csv")
df.head(2)
```

# Categorical

# Bar Graph

```python
sns.countplot(data=df,x="neighbourhood_group")
```

```python
sns.countplot(data=df,x="room_type")
```

```python
sns.countplot(data=df,x="neighbourhood_group",hue="room_type")
#In Seaborn, the hue parameter is used to add a second categorical variable that colors (splits) the bars, lines, or points in a plot.
```

# Pie Chart

```python
df["room_type"].value_counts().plot(kind="pie", autopct="%.2f")
#autopct="%.2f"  is used to format the numbers (percentages) displayed on each slice of the pie.
```

```python
df["room_type"].value_counts().plot(kind="bar")
```

```python
df["neighbourhood_group"].value_counts().plot(kind="pie", autopct="%.2f")
```
