# 1. Importing matplotlib and some necessary libraries
If we want to plot within an IPython notebook we can use %matplotlib. It comes with two different commands

%matplotlib notebook for interactive plots
%matplotlib inline for static plots
We will generally use %matplotlib inline

```python
import matplotlib as mp
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

1.1 Generating Sin and Cos Graph using Matplotlib

```python
x=np.linspace(0,3*np.pi,10)
sin_y=np.sin(x)
print(x)
print(sin_y)
plt.plot(x,sin_y)
plt.title('Sin wave')
plt.show()
```
