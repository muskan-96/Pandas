```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://quotes.toscrape.com/')
soup =BeautifulSoup(res.text,'html.parser')
#html.parser parses this HTML and converts it into a tree of elements so you can easily navigate, search, and extract data.
```

```python
soup.find('span',class_='text').text
```

```python
print(soup.find('span',class_='text').text[1:-1])
```

```python
quotees=[]
for quotes in soup.find_all('span',class_='text'):
    quotees.append(quotes.text[1:-1])
    
```
